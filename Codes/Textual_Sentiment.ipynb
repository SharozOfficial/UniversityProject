# Remove URL

sub	- Replaces one or many matches with a string

[]	A set of characters	"[a-m]"	
\	Signals a special sequence (can also be used to escape special characters)	"\d"	
.	Any character (except newline character)	"he..o"	
^	Starts with	"^hello"	
Dollar($)	Ends with	"planet$"	
*	Zero or more occurrences	"he.*o"	
+	One or more occurrences	"he.+o"	
?	Zero or one occurrences	"he.?o"	
{}	Exactly the specified number of occurrences	"he.{2}o"	
|	Either or	"falls|stays"

```python
#remove url from the text
import re
def remove_url(text):
  #remove url from the text
    url = []
    text = re.sub("https?://[^\s]+", "[URL]", text)
    url.append(text)
    return text
```

```python
# remove user from the text
def remove_user(text):
  #replace user names having token [USER]  
  text= re.sub("@\w+", "[USER]", text)
  return text
```

# Remove emojis
re. compile() method is used to compile a regular expression pattern provided as a string into a regex pattern object ( re. Pattern )

```python
#remove emojis from the text
def remove_emojis(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

# Remove Punctuation

```python
#remove punctuations from the text
def remove_punc(text):
  #replace them with blank
  text= re.sub("[^\w\s]", "", text)
  return text
```

```python
# remove new line from the text
def remove_newline(text):
    text= re.sub("\n", " ", text)
    return text
```

# Language Translation to English

The most basic use of the Google Translate API is, of course, translating words or sentences from one language into another. To do so, we have to import the Translator class from googletrans module

```python
import pandas as pd
#from py_translator import Translator
translator = Translator()
```

```python
#function to translate the source language into english
def translate(row):
    translator = Translator()
    if str(row['language']) == 'en':
        return str(row['Text'])
    else:
        #Translate the text with source language other than english into english language
        try:
            english = translator.translate(str(row['Text']), src=row['language'], dest='en').text
            return english
        except:
            return ''
```
